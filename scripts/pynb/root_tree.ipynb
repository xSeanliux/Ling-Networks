```python
import treeswift
```

```python
tree_path = "/projects/illinois/eng/cs/warnow/zxliu2/Ling-Networks/outputs/inference_outputs-2.0/high_borrowing/COV/trees/high1-_1_1.tree"
schema = "newick"
outgroup = "og"
```

```python
tree = treeswift.read_tree(tree_path, schema=schema)
tree.resolve_polytomies()
outgroup_node = tree.find_node(label=outgroup)
tree.reroot(node=outgroup_node, length=1e-9)
```

```python
tree.draw(show_labels=True)
```
